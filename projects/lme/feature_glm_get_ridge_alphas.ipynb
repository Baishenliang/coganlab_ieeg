```python
import pandas as pd
import numpy as np
for elec_grp in ('Auditory','Sensorymotor','Motor','Delay_only'):
    for alignment in ('Aud','Go','Resp'):
        
        filename = f"results/{elec_grp}_{alignment}_All.csv"
        raw = pd.read_csv(filename)
        raw['vWM_lambda'] = raw['vWM_lambda'].replace(1e4, np.nan)
        raw['novWM_lambda'] = raw['novWM_lambda'].replace(1e4, np.nan) # Included for completeness, though it's likely not in this column
        df_filtered = raw[raw['time_point'] > 0]
    
        max_vWM_lambda = np.nanmedian(df_filtered['vWM_lambda'])
        max_novWM_lambda = np.nanmedian(df_filtered['novWM_lambda'])
    
        print(f"{elec_grp} in {alignment} Maximum vWM_lambda: {max_vWM_lambda}")
        print(f"{elec_grp} in {alignment} Maximum novWM_lambda: {max_novWM_lambda}")
```
